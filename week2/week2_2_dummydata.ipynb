# 【問題1】乱数の作成

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [-3, 0] #平均
cov = np.array([[1.0, 0.8], [0.8, 1.0]]) #共分散行行列

# 2次元正規分布の乱数500個生成する
values = np.random.multivariate_normal(mean, cov, 500)
# print(values[:10, ])
# print(values.ndim)
# print(values.shape)

# シード値を0に設定する
np.random.seed(0)
```

# 【問題2】散布図による可視化

```python
# 散布図
plt.figure(dpi=90)
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(values[:, 0], values[:, 1], s=15, c='r')
plt.show()
```

# 【問題3】ヒストグラムによる可視化

```python
plt.figure(dpi=90)
plt.title("histgram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.xlim(-5.5, 2.5)
plt.hist(values[0:, 0:1], bins=50, color='r')
plt.show()

```

```python
plt.figure(dpi=90)
plt.title("histgram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.xlim(-5.5, 2.5)
plt.hist(values[0:, 1:2], bins=50, color='r')
plt.show()
```

# 【問題4】データの追加

```python
mean_1 = [-3, 0] #平均
cov = np.array([[1.0, 0.8], [0.8, 1.0]]) #共分散行行列

# 2次元正規分布の乱数500個生成する
values_1 = np.random.multivariate_normal(mean_1, cov, 500)

mean_2 = [0, -3] #平均

# 2次元正規分布の乱数500個生成する
values_2 = np.random.multivariate_normal(mean_2, cov, 500)

plt.figure(dpi=90)
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(values_1[:, 0], values_1[:, 1], s=15, label="0", c='r' )
plt.scatter(values_2[:, 0], values_2[:, 1], s=15, label="1", c='m')
plt.legend(loc=3)
```

# 【問題5】データの結合

```python
join_ndarray = np.concatenate([values_1, values_2]) #問題1，4で作成した配列を結合
print("2つの配列を結合すると\n{}".format(join_ndarray))
print("配列の形は{}\n".format(join_ndarray.shape))
```

# 【問題6】ラベル付け

```python
zero = np.zeros(500).reshape(500, 1) #0を500個並べた配列
one = np.ones(500).reshape(500, 1)#1を500個並べた配列
join_1 = np.hstack((values_1, zero)) #問題１の配列に0をラベル付け
join_2 = np.hstack((values_2, one)) #問題4の配列に1をラベル付け
join_all = np.vstack((join_1, join_2)) #ラベル付けしたもの同士を水平方向に結合
print("ラベル付けしたものを結合すると\n{}".format(join_all))
print("ラベル付けした配列のサイズは：{}".format(join_all.shape))
```
