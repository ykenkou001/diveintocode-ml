```python
def gini(y):
    '''
    '''
    #クラスのユニーク値
    c = np.unique(y)
    #クラスの数
    k = len(y)
    
    sigma = 0
    for i in range(len(c)):
        w = np.count_nonzero(y == [i])
        l = w/k
        sigma += l**2
    i_t = 1 - sigma

    return i_t
```

```python
#情報利得を求める関数
def information_gain(y, left, right):
    
    #left,rightはインデックス？
    info_gain = gini(y)- (left/len(y))*gini(left) - (right/len(y)*gini(right)
    
    return info_gain
```
