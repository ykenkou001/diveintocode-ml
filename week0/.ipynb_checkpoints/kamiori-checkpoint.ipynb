```python
#問題1

""""
紙を1回折った時の厚さを計算するコード
"""

THICKNESS = 0.00008

ori_time = 2 ** 43

folded_thickness = THICKNESS * ori_time

print("厚さ： {}メートル".format(folded_thickness))
```

```python
#問題2

THICKNESS = 0.00008

ori_time = 2 ** 43

folded_thickness = THICKNESS * ori_time

print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
```

```python
#問題３

THICKNESS = 0.00008

for a in range(43):
    THICKNESS *= 2
    
print("厚さ： {}メートル".format(THICKNESS))
```

```python
#問題４-１

import time
start = time.time()
#####
THICKNESS = 0.00008

ori_time = 2 ** 43

folded_thickness = THICKNESS * ori_time

# print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
#問題４-２

import time
start = time.time()
#####
THICKNESS = 0.00008
for a in range(43):
    THICKNESS *= 2
# print("厚さ： {}メートル".format(THICKNESS))
####
elapsed_time = time.time() - start

print("time : {}[s]".format(elapsed_time))
```

```python
#問題５

list = []
THICKNESS = 0.00008
list.append(THICKNESS)
folded_thickness = THICKNESS

for i in range(43):
    folded_thickness *= 2
    list.append(folded_thickness)

print(list)
print(len(list))
```

```python
#問題５（別バージョン）

list = []
THICKNESS = 0.00008
list.append(THICKNESS)

for i in range(43):
    THICKNESS *= 2
    list.append(THICKNESS)
    
print(list)
print(len(list))
```

```python
#問題　６

"""
#グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(list) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

グラフに対する考察：折る回数が３０回後半から爆発的に厚みが増していく。

```python
#問題７　グラフのカスタマイズ-1
"""
赤い折れ線グラフを表示する。
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(list, color='red')
plt.show()
```

```python
"""
軸の値を大きく表示する。
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(list)
plt.show()
```

```python
"""
線種を変える。。
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(list, linestyle = "dashed")
plt.show()
```
