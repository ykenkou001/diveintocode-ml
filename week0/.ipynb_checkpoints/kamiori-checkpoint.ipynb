# 【問題1】べき乗の算術演算子を使用して作成

```python
"""
紙を1回折った時の厚さを計算するコード
"""

THICKNESS = 0.00008

folded_thickness = THICKNESS*2**43


print("厚さ： {}メートル".format(folded_thickness))
```

# 【問題2】単位の変換

```python
"""
紙を1回折った時の厚さを計算するコード
"""

THICKNESS = 0.00008

folded_thickness = THICKNESS*2**43

# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}キロメートル".format(folded_thickness/1000))
```

# 【問題3】for文を使用して作成

```python
THICKNESS = 0.00008 #紙を折る前の紙の厚さ
n_folded_thickness = [] # 43回紙を折った時のリスト

for i in range(43):
    THICKNESS *= 2
    n_folded_thickness.append(THICKNESS)
    
# print(n_folded_thickness)
# print(len(n_folded_thickness))
print("厚さ： {}メートル".format(folded_thickness))
```

# 【問題4】計算時間の比較

### べき乗を使用したバージョン 

```python
import time
start = time.time()
#####
THICKNESS = 0.00008

folded_thickness = THICKNESS*2**43

#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

# 【問題5】リストへの保存

```python
THICKNESS = 0.00008 #紙を折る前の紙の厚さ
n_folded_thickness = [] # 43回紙を折った時のリスト
n_folded_thickness.append(THICKNESS)

for i in range(43):
    THICKNESS *= 2
    n_folded_thickness.append(THICKNESS)
    
# print(n_folded_thickness)
print(len(n_folded_thickness))
# print("厚さ： {}メートル".format(folded_thickness))
```

# 【問題6】折れ線グラフの表示

```python
"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(dpi=100)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(n_folded_thickness) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

# 【問題7】グラフのカスタマイズ

```python
"""
マゼンダの折れ線グラフを表示する。
"""
plt.figure(dpi=100)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(n_folded_thickness, color='m')
plt.show()
```

```python
"""
破線の折れ線グラフを表示する。
"""
plt.figure(dpi=100)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(n_folded_thickness, ls='--', color='b')
plt.show()
```

```python
"""
折れ線グラフの線の太さを変える。
"""
plt.figure(dpi=100)
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(n_folded_thickness, lw=6, color='r')
plt.show()
```
