```python
# 体積を求める関数を定義する。
def ask_v(r):
    return ((4 * 3.14 * r**3) / 3)

kuri_v = ask_v(2.5e-5) # 栗饅頭の体積(km***3)
solar_sys_v = ask_v(5.954e+9)# 太陽系の体積(km***3)

n_kuri_v = []

while True:
    kuri_v *= 2
    n_kuri_v.append(kuri_v)
    if n_kuri_v[-1] > solar_sys_v:
        break

n_times = 5 * len(n_kuri_v)        
        
print("{}分後に栗饅頭が太陽系を覆う".format(n_times))
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'Noto Sans CJK JP']

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("栗饅頭の体積の変化")
plt.xlabel("バイバインを振りかけた回数")
plt.ylabel("栗饅頭の体積の変化")
plt.plot(n_kuri_v)
plt.show()
```
