```python
#栗まんじゅうの半径：0.025ｍとする
#太陽系を球体と仮定し計算する。
#太陽から太陽系の最も外側の惑星海王星までの距離：4495000000000ｍとする
#太陽系について→　https://ja.wikipedia.org/wiki/%E6%B5%B7%E7%8E%8B%E6%98%9F

import math

# 体積を求める関数を定義する。
def ask_v(r):
    return (4/3 * math.pi * r**3)

# バイバインを振りかけた物体が何分後にある空間を一杯にするかを出力する関数を定義する。
def fill_in(n, m):
    
    first_v = ask_v(n) #物体の最初の体積
    space = ask_v(m) #ある空間の体積
    volume_up = [] #増加した物体の体積を格納するリスト
    volume_up.append(first_v) #物体の最初の体積をリストに格納

    while volume_up[-1] <= space:
        first_v *= 2
        volume_up.append(first_v)

    n_times = 5 * (len(volume_up) - 1)
    
    return n_times

n = 0.025 # 栗饅頭の半径
m = 4495000000000 # 太陽系の半径
        
print("{}分後に栗饅頭が太陽系を覆う".format(fill_in(n,m)))
```

```python
#栗饅頭の体積の変化を格納するリストを作成

n = 0.025 # 栗饅頭の半径
m = 4495000000000 # 太陽系の半径

first_v = ask_v(n) #栗饅頭の体積
space = ask_v(m) #太陽系の体積
volume_up = [] #増加した栗饅頭の体積の変化を格納するリスト
volume_up.append(first_v) #1つの栗饅頭の体積をリストに格納

while volume_up[-1] <= space:
    first_v *= 2
    volume_up.append(first_v)
```

```python

from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'Noto Sans CJK JP']

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(dpi=100)
plt.title("栗饅頭の体積の変化")
plt.xlabel("バイバインを振りかけた回数")
plt.ylabel("栗饅頭の体積の変化")
plt.plot(volume_up)
plt.show()
```
