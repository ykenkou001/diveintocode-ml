```python
import matplotlib.pyplot as plt
%matplotlib inline


# 定数の設定
GRAIN_START = 1
DAY = 100
MAG = 2 # 倍率

# 過程の値を格納する空のリスト
# リストは「その日にもらう米の数」と「その日までに累計で貰う米の数」の二つ
grain_receive = []
grain_total = []
days =[]


# 数の計算
# ループで処理
# 1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」を計算
for i in range(DAY):
   days.append(i + 1) # n日目は1日目から数えるため
   grain_receive.append(GRAIN_START * MAG ** i)
   grain_total.append(sum(grain_receive))

print(len(grain_receive))
print(len(grain_total))

# 折れ線グラフでプロット
plt.figure(figsize=(10, 5), dpi=200)
plt.title("Number of rice grains after N days")
plt.xlabel("days")
plt.ylabel("number of grains")
plt.plot(days, grain_receive, days, grain_total)  # ２つの軸で表したい
plt.show()
```
